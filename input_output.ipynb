```python
#to overcome this we use split it searches for space in string and creates it as list
#by default it splits " " and only for string

a,b=input().split()  #a,b =10 20.split()
print(a,b)           #a,b =["10", "20"]
print(a+b)           #a="10" b="20"  #type string
print(int(a+b)+1)    #it wont work for a+b as we need to covert into int first
```
