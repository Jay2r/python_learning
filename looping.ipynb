4. If the user enters correct password show success message and stop the loop 
5. If the user enters a erong password count the attempt 
6. If 3 wrong password happen, Block user and stop the loop 

2. Loop through transection Id - from 1000 to 1100

```python
# Build a Login System 

crct_password = "RajCloud1234"
count=0
for i in range(1,6):
    password = input("enter the password to login")
    if password == "":
        print("password cannot be empty please try again")
        continue
    elif password == crct_password:
        print("you have successfully logged in!!")
        break
    elif password != crct_password:
        print("please enter correct password")
        count+=1
        print(f"attempts left:{5-count}")
        if count == 5:
            print("max attempts reached you are blocked")
            break    

```
