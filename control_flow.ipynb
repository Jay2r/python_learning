```python
age = int(input("Enter your age"))
citizen = bool(input("if you are indian type yes & if ur not indian just press enter button"))

if age>=18:
    print("you are above 18")
    if citizen==True:
        print("you are indian and you can vote")
    else:
        print("you must be indain citizen")  
else:
    print("you are not eligible to vote")          
```

```python
is_logged_in=True
account_active=True
item_in_stock=True
payment_successful=True
delivery_serviceable=False

if is_logged_in:
    print("User is logged in")
    if account_active:
        print("Account is active")
        if item_in_stock:
            print("Item is in stock")
            if payment_successful:
                print("Payment successful")
                if delivery_serviceable:
                    print("Order placed successfully!")
                else:
                    print("Delivery not available to your location.")
            else:
                print("Payment failed. Please try again.")
        else:
            print("Item is out of stock.")
    else:
        print("Account is inactive. Please contact support.")
else:
    print("Please log in to continue.")

```
