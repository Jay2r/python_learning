```python
#PROGRAM TO CHECK WHETHER A NUMBER IS PRIME OR NOT

N = int(input("Enter a number"))
is_prime = True

for i in range(2,N):

    if(N%i == 0):
        is_prime = False
        break   
if is_prime == True:    
    if(N%2 == 0):
        print(f"the number is:{N} and it is prime even")
    else: 
        print(f"the number is:{N} and it is prime odd")  
else:
    print("it is non prime")
```

```python
# PROGRAM TO CHECK PRIME NUMBERS FROM 2 TO N

N = int(input("Enter a number"))
c = N/2
if N<=1:
   print(f"it is not a prime number:{N}")
for num in range(2,N+1):
    is_prime = True
    for i in range(2,num):
        if(num%i ==0):
            is_prime = False
            break
    if is_prime == True: 
       if(num%2 == 0):
        print(f"it is prime even and the number is:{num}")    
       else: 
        print(f"it is prime odd and the number is:{num}")  
    elif num%2 == 0:
         if num > c:
            print(f"it is non prime even upper half:{num}")      
         else:
            print(f"it is non prime even lower half:{num}")
    else:
       if num%3==0:
          print(f"the number is a multiple of 3 and it is non prime odd:{num}")
       else:
          print(f"the number is non prime odd:{num}")           
```
