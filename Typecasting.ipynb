What is Type Casting?

Type casting means converting one data type into other data type.

"10" (string). --> 10 int 
10 (int) ------> 10.0 float 

Why Type casting Needed?

- Input() always return string 
- Operations between diffrent datatype may cause errors 


Type of Casting 

1. Implicit - Python converts automatically 

2. Explicit - Programmer converts manually 

What is Implicit Type Casting?
Simple Definition

Implicit type casting is when Python automatically converts one data type into another without the programmer explicitly asking for it.

Python does this ONLY when it is safe.

a = 10      # int
b = 2.5     # float

c = a + b
print(c, type(c))


Python automatically converted 10 (int) → 10.0 (float)



```python
a=10
b=20.5
c=a+b
print(c)
print(type(c))
```

Why Python perform Implicit Casting? 
1. Prevent data loss 
2. Maintain Mathamatical Correctness 

    If conversion might 
    Lose precision

    Be Ambiguous 

Lower ------> Higher Data type

Rule -2

What is Explicit Type Casting?
Simple Definition

Explicit type casting is when the programmer manually converts one data type into another using built-in functions.

Python will never guess your intent — you must tell it explicitly.

x = "10"
y = int(x)

Here:

"10" → string

int(x) → explicitly convert to integer

Why Explicit Casting is Needed

Python requires explicit casting when:

Conversion may cause data loss

Conversion is ambiguous

Conversion changes meaning

"10" + 5   # ❌ ambiguous


int() CASTING (MOST USED)
What int() Does

Converts a value into a whole number.

int("10")      # 10
int(10.9)      # 10
int(True)      # 1
int(False)     # 0


float() CASTING
What float() Does

Converts value into decimal number.

float("10")     # 10.0
float("10.5")   # 10.5
float(5)        # 5.0
float(True)     # 1.0


str() CASTING
What str() Does

Converts value into string (text).

x = 100
y = str(x)
